```python
# Replace 'your_file.csv' with the actual name of your file
df = pd.read_csv('MonthlyRemoteSensingData3_Mtuba.csv', delimiter= ';', skipinitialspace=True)

# Display the first few rows to understand the structure
print(df.head())
```

```python
# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(df)

# Convert the Date column to the desired format (YYYY-MM-DD)
df["date"] = pd.to_datetime(df["date"], format="%Y/%m/%d").dt.strftime("%Y-%m-%d")
df.head()
```

```python
# Set the 'Date' column as the index
df.set_index('date', inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your DataFrame and it has the necessary columns:EVI'

# Create a figure and an axes object (ax)
fig, ax = plt.subplots()  # Create the subplot before plotting

# Plot 'EVI' as line charts
df['SAVI'].plot(kind='line', ax=ax, color='brown', label='SAVI')

# Add a title and labels
ax.set_title('SAVI Trend over 5 Year Time Series')
ax.set_xlabel('Dates')
ax.set_ylabel('INDEX Value')

# Add a legend to differentiate the lines and bars
ax.legend()

# Hide the top and right spines for a cleaner look
ax.spines[['top', 'right']].set_visible(False)

# Make x-axis labels slanted (e.g., 45 degrees)
plt.xticks(rotation=45, ha='right')  # ha='right' aligns labels to the right

# Show the plot
plt.tight_layout()
plt.show()
```

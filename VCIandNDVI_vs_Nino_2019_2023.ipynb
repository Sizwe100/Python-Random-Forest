```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Simulated dataset for demonstration purposes
np.random.seed(42)
dates = pd.date_range(start="2019-01-01", periods=60, freq='ME')  # 'ME' stands for end of the month

df = pd.DataFrame({
    'Date': dates,
    'NDVI_LC': np.random.uniform(0.3, 0.8, len(dates)),
    'VCI_LC': np.random.uniform(20, 100, len(dates)),
    'NDVI_UVS': np.random.uniform(0.3, 0.8, len(dates)),
    'VCI_UVS': np.random.uniform(20, 100, len(dates))
})

# Convert 'Date' to datetime format and set as index
df.set_index('Date', inplace=True)

# Define scientific plotting style
plt.style.use('ggplot')  # Alternative built-in style

# Create a directory for saving plots
save_path = "plots"
os.makedirs(save_path, exist_ok=True)  # Create folder if it doesn't exist

# Function to create a thesis-ready plot
def plot_variable(df, column, color, title, ylabel, filename):
    fig, ax = plt.subplots(figsize=(10, 6))
    df[column].plot(ax=ax, color=color, linewidth=2)

    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.tick_params(axis='x', rotation=45)

    # Hide unnecessary spines
    ax.spines[['top', 'right']].set_visible(False)

    plt.tight_layout()

    # Save plot in the 'plots' directory
    save_location = os.path.join(save_path, filename)
    plt.savefig(save_location, dpi=300)

    plt.show()

# Generate and save the four separate plots
plot_variable(df, 'NDVI_LC', 'darkred', 'NDVI Trends at LaChagra', 'NDVI Value', 'NDVI_LC_plot.png')
plot_variable(df, 'VCI_LC', 'black', 'VCI Trends at LaChagra', 'VCI Value', 'VCI_LC_plot.png')
plot_variable(df, 'NDVI_UVS', 'darkblue', 'NDVI Trends at UVS', 'NDVI Value', 'NDVI_UVS_plot.png')
plot_variable(df, 'VCI_UVS', 'grey', 'VCI Trends at UVS', 'VCI Value', 'VCI_UVS_plot.png')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Simulated dataset for demonstration purposes
np.random.seed(42)
dates = pd.date_range(start="2019-01-01", periods=60, freq='ME')  # 'ME' stands for end of the month

df = pd.DataFrame({
    'Date': dates,
    'NDVI_LC': np.random.uniform(0.3, 0.8, len(dates)),
    'VCI_LC': np.random.uniform(20, 100, len(dates)),
    'NDVI_UVS': np.random.uniform(0.3, 0.8, len(dates)),
    'VCI_UVS': np.random.uniform(20, 100, len(dates))
})

# Convert 'Date' to datetime format and set as index
df.set_index('Date', inplace=True)

# Define anomalies (highlighting values below 25th percentile or above 75th percentile)
def detect_anomalies(series):
    q25, q75 = np.percentile(series, [25, 75])
    return (series < q25) | (series > q75)  # Boolean mask for anomalies

# Create a directory for saving plots
save_path = "plots"
os.makedirs(save_path, exist_ok=True)  # Create folder if it doesn't exist

# Use the default Matplotlib style for a clean look
plt.style.use('default')

# Create a 2x2 grid of plots
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # Adjusted size for half-page fit

# Define plot details
plot_details = [
    ('NDVI_LC', 'darkred', '(a) NDVI Trends at LaChagra'),
    ('VCI_LC', 'black', '(b) VCI Trends at LaChagra'),
    ('NDVI_UVS', 'darkblue', '(c) NDVI Trends at UVS'),
    ('VCI_UVS', 'grey', '(d) VCI Trends at UVS')
]

# Set Times New Roman font
plt.rcParams['font.family'] = 'Times New Roman'

# Loop through and plot each variable
for ax, (column, color, title) in zip(axes.flatten(), plot_details):
    df[column].plot(ax=ax, color=color, linewidth=1.5, marker='o', markersize=3)

    # Highlight anomalies
    anomalies = detect_anomalies(df[column])
    ax.scatter(df.index[anomalies], df[column][anomalies], color='red', zorder=3, label='Anomalies')

    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.set_xlabel('Date', fontsize=10)
    ax.set_ylabel('Index Value', fontsize=10)
    ax.tick_params(axis='x', rotation=45, labelsize=8)
    ax.tick_params(axis='y', labelsize=8)
    ax.spines[['top', 'right']].set_visible(False)
    ax.set_facecolor('white')

    ax.legend(loc='upper right', fontsize=8)  # Add legend for anomalies

# Adjust layout for compactness
plt.tight_layout()

# Save the combined figure
save_location = os.path.join(save_path, "Combined_VCI_NDVI_Plots_Anomalies.png")
plt.savefig(save_location, dpi=300, bbox_inches='tight')

# Show the final combined plot
plt.show()
```
